```python
import pandas as pd
import numpy as np

# Define the updated departments and their specific skills
departments = {
    "Data Science": ["ML", "DL", "Python", "Statistics", "Data Wrangling", "R"],
    "Software Engineering": ["Java", "C++", "Python", "Algorithms", "Data Structures", "Git"],
    "Human Resources": ["Recruitment", "Training", "Communication", "Conflict Resolution", "Employee Engagement", "HR Policies"],
    "Marketing": ["SEO", "Content Creation", "Analytics", "Social Media", "Brand Management", "Market Research"],
    "Finance": ["Accounting", "Financial Analysis", "Excel", "Budgeting", "Taxation", "Investment Analysis"]
}

# List of all possible skills
all_skills = ["ML", "DL", "Python", "Statistics", "Data Wrangling", "R",
              "Java", "C++", "Algorithms", "Data Structures", "Git",
              "Recruitment", "Training", "Communication", "Conflict Resolution",
              "Employee Engagement", "HR Policies",
              "SEO", "Content Creation", "Analytics", "Social Media", "Brand Management", "Market Research",
              "Accounting", "Financial Analysis", "Excel", "Budgeting", "Taxation", "Investment Analysis"]

# Initialize the employee ID counter
employee_id_counter = 1

# Generate the dataset
np.random.seed(42)
data = []

for _ in range(150):
    # Randomly select a department
    department = np.random.choice(list(departments.keys()))

    # Initialize the skill ratings to zero
    skill_ratings = {skill: 0 for skill in all_skills}

    # Assign ratings for the department-specific skills
    for skill in departments[department]:
        skill_ratings[skill] = np.random.randint(1, 11)  # Ratings between 1 and 10

    # Create the row
    row = {"Employee ID": employee_id_counter, "Department": department, **skill_ratings}
    data.append(row)

    # Increment the employee ID counter
    employee_id_counter += 1

# Create a DataFrame
df = pd.DataFrame(data)

# Save to CSV
file_path = "employee_skills_ratings_V3.csv"
df.to_csv(file_path, index=False)

print(f"File saved as: {file_path}")
df
```
